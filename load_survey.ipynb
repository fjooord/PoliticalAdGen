```python
import PyPDF2
import re
```

```python
def extract_text_from_pdf(file_path, start_page, end_page):
    # Open the PDF file in read-binary mode
    with open(file_path, 'rb') as file:
        # Create a PDF file reader object
        reader = PyPDF2.PdfReader(file)

        # Check the total number of pages in the PDF
        num_pages = len(reader.pages)

        # Check if the provided page range is valid
        if start_page < 1 or end_page > num_pages or start_page > end_page:
            raise ValueError("Invalid page range")

        # Adjust for 0-based index
        start_page -= 1
        end_page -= 1

        # List to store the text of each page
        text_list = []

        # Read the text from each page
        for page in range(start_page, end_page + 1):
            text = reader.pages[page].extract_text().strip()
            
            text_list.append(text)

        return text_list
def extract_questions(text_array):
    # Use re.DOTALL flag to make . match any character including newline
    pattern = re.compile(r'\b\d+\.\s.*?\?\s', re.DOTALL)
    questions = []
    for page_text in text_array:
        matches = re.findall(pattern, page_text)
        questions.extend([question.replace('\n', ' ').strip() for question in matches])
    return questions

# Usage
file_path = '2022-mountain-states-poll-report.pdf'
start_page = 4
end_page = 31
text_array = extract_text_from_pdf(file_path, start_page, end_page)

for t in text_array:
    print(t)


for t in extract_questions(text_array):
    print(t)
```
